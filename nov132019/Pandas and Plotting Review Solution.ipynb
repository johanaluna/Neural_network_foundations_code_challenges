## Create a Crosstab

```python
cross = pd.crosstab(index = pd.cut(df['Cement (component 1)(kg in a m^3 mixture)'], 10),
                    columns = df['Age (day)'],
                    values = df['Concrete compressive strength(MPa, megapascals) '],
                    aggfunc = 'mean')

cross
```

## Create distribution plots for each feature

## Fit a regression model and extract feature importances 

```python
x = df.drop(columns = 'Concrete compressive strength(MPa, megapascals) ')
y = df['Concrete compressive strength(MPa, megapascals) ']

from sklearn.model_selection import train_test_split

x_test, x_train, y_test, y_train = train_test_split(x, y, test_size=.2, random_state=42)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV

boost = make_pipeline(StandardScaler(),
                      GradientBoostingRegressor())


params = {'gradientboostingregressor__max_depth':[2, 5, 10, 20]}

gs_boost = GridSearchCV(boost, param_grid=params, scoring='r2', n_jobs=-1, cv=10, verbose=10, return_train_score=False)

gs_boost.fit(x, y)

gs_boost = gs_boost.best_estimator_
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

r2_score(gs_boost.predict(x_test), y_test), mean_absolute_error(gs_boost.predict(x_test), y_test)
```

```python
feature_importance = gs_boost.steps[1][1].feature_importances_

feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5

plt.subplot(1, 2, 2)
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, df.columns[sorted_idx])
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
plt.show()
```
