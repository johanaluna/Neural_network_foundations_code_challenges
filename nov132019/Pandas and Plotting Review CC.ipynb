## Preform a Groupby

```python
df_gropup= df.groupby('Age (day)').mean()
df_gropup
```

## Create a Crosstab

```python
df_crosstab=pd.crosstab(df['Age (day)'], df['Concrete compressive strength(MPa, megapascals) ']).T
```

```python
df_crosstab
```

## Create a heatmap with your crosstab

```python

import seaborn as sns

sns.set()

ax = sns.heatmap(df_crosstab,cmap="YlGnBu", annot=True, cbar=False)
```
